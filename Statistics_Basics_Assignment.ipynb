1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales.

ans- Data Types:

Qualitative (Categorical): Describes qualities or characteristics.

Nominal: Data that can only be categorized, with no inherent order.
Examples: Gender (Male, Female), Eye Color (Blue, Brown, Green), Marital Status (Single, Married, Divorced)
Ordinal: Data that can be categorized and ranked, but the differences between categories may not be equal.
Examples: Educational Level (High School, Bachelor's, Master's), Satisfaction Rating (Poor, Fair, Good, Excellent)
Quantitative (Numerical): Represents measurable quantities.

Interval: Data with a meaningful order and equal intervals between values, but no true zero point.
Examples: Temperature (Celsius, Fahrenheit), Calendar Years
Ratio: Data with a meaningful order, equal intervals, and a true zero point (indicating the absence of the quantity).
Examples: Height, Weight, Income, Age

2. What are the measures of central tendency, and when should you use each? Discuss the mean, median, and mode with examples and situations where each is appropriate.

ans- Measures of Central Tendency:

Mean: The average of a dataset. Use when data is roughly symmetric and not heavily influenced by outliers.

Example: Average income in a city.
Median: The middle value when data is ordered. Use when data is skewed or has outliers.

Example: Median house price in a neighborhood (outliers like mansions can skew the mean).
Mode: The most frequent value. Use to identify the most common category or value.

Example: Most popular car brand in a region.

3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?

ans- Dispersion:

Measures how spread out the data points are.
Variance: The average squared difference of each data point from the mean. It gives a measure of variability in units squared.
Standard Deviation: The square root of the variance. It provides a measure of variability in the same units as the data, making it easier to interpret.

4. What is a box plot, and what can it tell you about the distribution of data?

ans- A box plot (also known as a box-and-whisker plot) visually represents the five-number summary of a dataset: minimum, first quartile (Q1), median, third quartile (Q3), and maximum.
It shows the distribution of data, including potential skewness, outliers, and the spread of the middle 50% of the data (interquartile range).

5. Discuss the role of random sampling in making inferences about populations.

ans- Random sampling allows researchers to draw conclusions about a population based on a smaller sample.
By randomly selecting individuals from the population, we ensure that the sample is representative and reduces bias.
This allows for more accurate and reliable inferences about the population characteristics.

6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?

ans- Skewness measures the asymmetry of a distribution.
Positive Skew (Right-skewed): The tail of the distribution extends to the right. Mean > Median
Example: Income distribution (many people have lower incomes, a few have very high incomes).
Negative Skew (Left-skewed): The tail of the distribution extends to the left. Mean < Median
Example: Test scores where most students score high, but a few score low.
Skewness affects the choice of central tendency measure and can influence the interpretation of statistical results.

7. What is the interquartile range (IQR), and how is it used to detect outliers?

ans- Interquartile Range (IQR): The difference between the third quartile (Q3) and the first quartile (Q1). It represents the spread of the middle 50% of the data.
Outlier Detection: Values below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR are often considered potential outliers.

8. Discuss the conditions under which the binomial distribution is used.

ans- The binomial distribution models the number of successes in a fixed number of independent trials, where each trial has only two possible outcomes (success or failure).
Conditions:
Fixed number of trials (n)
Each trial is independent
Two possible outcomes for each trial
Probability of success (p) remains constant across trials

9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).

ans- Normal Distribution (Gaussian Distribution):
Symmetrical, bell-shaped curve
Mean, median, and mode are equal
Defined by its mean (μ) and standard deviation (σ)
Empirical Rule (68-95-99.7 Rule):
Approximately 68% of the data falls within one standard deviation of the mean.
Approximately 95% of the data falls within two standard deviations of the mean.
Approximately 99.7% of the data falls within three standard deviations of the mean.

10. Provide a real-life example of a Poisson process and calculate the probability for a specific event.

ans- Poisson Process: Describes the number of events occurring within a fixed interval of time or space, given that these events happen randomly and independently with a constant average rate.
Example: The number of customers arriving at a bank per hour.
To calculate the probability of a specific number of events, we would use the Poisson probability formula, which involves the rate parameter.

11. Explain what a random variable is and differentiate between discrete and continuous random variables.

ans- Random Variable: A variable whose value is a numerical outcome of a random phenomenon.
Discrete Random Variable: Can take on a countable number of distinct values (e.g., number of heads in coin flips, number of cars passing by in an hour).
Continuous Random Variable: Can take on any value within a given interval (e.g., height, weight, temperature).
